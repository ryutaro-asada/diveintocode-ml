```python
def compute_gradient(function = 'function1', X_range=(-50, 50.1, 0.1)):
    X_array = np.arange(X_range[0], X_range[1], X_range[2])
    X_array = X_array.reshape(X_array.shape[0], 1)
#     xr_ndarray = x_ndarray.reshape(1001,1 )
    if function == None:
        print('you must put function in')
        
    if function == 'function1':
        y_array = X_array/2 + 1 # Xのndarrayを２で割って１足す
        
    if function == 'function2':
        y_array = X_array**2   # Xのndarrayを２二乗
        
    if function == 'function3':
        y_array = (2*X_array**2) + 2**X_array
        
    if function == 'function4':
        y_array = np.sin(X_array**(1/2))
        
    x_y_array = np.concatenate((X_array, y_array), axis=1)
    
    d_xr_ndarray = X_array[1:] - X_array[:-1] #Xの変化量
    # print(d_xr_ndarray)
    d_yr_ndarray = y_array[1:] - y_array[:-1] #Xの変化量
    # print(d_yr_ndarray)
    koubai_ndarray = d_yr_ndarray/d_xr_ndarray
    return x_y_array, koubai_ndarray
    
compute_gradient('function1', X_range = (-50, 50, 0.1))
```

```python
X_y_array, koubai_ndarray = compute_gradient('function1', (-50, 50, 0.1))

plt.plot(X_y_array[:, 0][:-1], koubai_ndarray, label = 'grad')
plt.plot(X_y_array[:, 0], X_y_array[:, 1], label = 'x_y_line')
plt.xlabel('point')
plt.ylabel('hight')
plt.legend()
plt.show()
```
